#### Plot RGB Histogram

```python
# yellow banana
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("yellow.jpg")
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr, color = col, )
    
    # set the x, y limits
    plt.xlim([0,250])
    plt.ylim([0,625])
plt.show()
```

## Checkpoint 4
### Gray Scale Image

```python
import numpy 
import cv2
from matplotlib import pyplot as plt

my_id = "6610110341"
print(f"My ID is:{my_id}")

image = cv2.imread("green.jpg")

# rows, columns, and channels
image_size = image.shape
width = image_size[0]
high = image_size[1]
# uint8 --> Non-negative integers ranging from 0 to 255 
image_gray = (numpy.uint8(numpy.zeros((width, high))))

cv2.imshow("image_gray", image_gray)
for i in range(width):
    for j in range(high):
        image_gray[i,j] = (float(image[i,j][0])+float(image[i,j][1])+float(image[i,j][2]))/3.0

cv2.imshow("image_gray", image_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread("green.jpg")

image_size = image.shape
width = image_size[0]
high = image_size[1]

for i in range(width):
    for j in range(high):
        #        b, g, r
        if (image[i,j,1])<250:
            image[i,j,0] = image[i,j,0]+50



cv2.imshow("adding_image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
