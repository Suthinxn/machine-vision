### Image processing Gray to Binary

```python
import numpy
import cv2
from matplotlib import pyplot as plt

my_id = "6610110341"
print(f"My ID is:{my_id}")

image = cv2.imread("green.jpg")

image_size = image.shape
width = image_size[0]
high = image_size[1]
image_gray = numpy.uint8(numpy.zeros((width,high)))

for i in range(width):
    for j in range(high):
        image_gray[i,j] = numpy.uint8((float(image[i,j,0]) + float(image[i,j,1]) + float(image[i,j,2]))/3.0)

print(image_gray)

image_binary = numpy.uint8(numpy.zeros((width,high)))
for i in range(width):
    for j in range(high):

        if image_gray[i,j]>200:
            image_binary[i,j] = 255
        else:
            image_binary[i,j] = 0
            

cv2.imshow("Gray image", image_gray)
cv2.imshow("Binary image", image_binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##  Checkpoint 7
---
### background subtraction

```python
import cv2
import numpy

background = cv2.imread('background.jpg', cv2.IMREAD_GRAYSCALE)
current_image = cv2.imread('object.jpg', cv2.IMREAD_GRAYSCALE)

if background.shape != current_image.shape:
    current_image = cv2.resize(current_image, (background.shape[1], background.shape[0]))


height, width = background.shape

image_blank = numpy.zeros((height, width), dtype=numpy.uint8)

for i in range(height):
    for j in range(width):
        # result = |x - y|
        image_blank[i][j] = abs(int(background[i][j]) - int(current_image[i][j]))


cv2.imshow('Difference', image_blank)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
